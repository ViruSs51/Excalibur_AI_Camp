```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]

plt.plot(x, y)
plt.xlabel('Aceasta este axa X')
plt.ylabel('aceasta este axa Y')
plt.title('titlu pentru grafic')
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.exp(x)

plt.plot(x, y)
plt.show()

plt.plot(x, y)
plt.xscale('log')
plt.show()

plt.plot(x, y)
plt.yscale('log')
plt.show()
```

```python
plt.plot(x, y)
plt.show()

plt.plot(x, y)
plt.xlim(8, 10)
plt.ylim(5000, 10**6)
plt.show()

plt.plot(x, y)
plt.xlim(8, 10)
plt.ylim(5000, 10**6)
plt.xticks([8.00, 9.00, 10.00])
plt.yticks([0.0, 500000.0, 1000000.0])
plt.show()
```

```python
x = np.linspace(0,10, 100)
y1 = x
y2 = x**2

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, np.sqrt(x))

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, np.sqrt(x))

plt.show()
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
c = np.random.randn(100)

plt.scatter(x, y, c=x+y, cmap='viridis')

plt.colorbar()
plt.show()
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)
c = np.random.randn(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(x, y, z, c=c)
plt.xlabel('x label')
plt.title('3d graf')
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y1 = [100, 200, 300, 400, 500]
y2 = [50, 100, 150, 200, 250]

fig, ax1 = plt.subplots()

ax1.plot(x, y1)
ax1.set_ylabel('One axs')

ax2 = ax1.twinx()
ax2.plot(x, y2)
ax2.set_ylabel('Two axs')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y1, y2)
plt.show()
```
